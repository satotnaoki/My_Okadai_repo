```python
def eea(u, v):
	u1 = 1
	u2 = 0
	u3 = u
	v1 = 0
	v2 = 1
	v3 = v
	while v3 != 0:
		q = u3 / v3
		t1 = u1 - q * v1
		t2 = u2 - q * v2
		t3 = u3 - q * v3
		u1 = v1
		u2 = v2
		u3 = v3
		v1 = t1
		v2 = t2
		v3 = t3
	return u1, u2, u3
```

```python
from collections import deque

r_array = deque()
q_array = deque()
s_array = deque()
t_array = deque()
i_array = deque()

def eea3_book (r0, r1):
    global r_array
    global q_array
    global s_array
    global t_array
    global i_array
    
    s0 = 1
    s1 = 0
    s_array.append(s0)
    s_array.append(s1)
    
    t0 = 0
    t1 = 1
    t_array.append(t0)
    t_array.append(t1)
    
    r_array.append(r0)
    r_array.append(r1)

    i = 1
    i_array.append(0) #i[0]
    i_array.append(1) #i[1]
    
    q_minus1 = "-"
    q_array.append(q_minus1)
    q0 = "-"
    q_array.append(q0)
    
    while (r_array[i] != 0):
     i = i + 1                                               #1.1
     i_array.append(i) #starts at i[1], always i = 2
     r_incr = r_array[i-2] % r_array[i-1]                    #1.2
     r_array.append(r_incr)                                  #1.2
     q_decr = (r_array[i-2] - r_array[i])/r_array[i-1]       #1.3
     q_array.append(q_decr)                                  #1.3
     q_array[i] = q_decr                                     #1.3 
        #Very important => Position 0, q(-1) = - since r(-2) doesn't exist
        # Position 1, q(0) = - since r(-1) doesn't exist
        # Position 2, q(1) = 5 since r(0) exists 
     s_incr = s_array[i-2] - (q_array[i]*s_array[i-1])     #1.4
        #Very important => Array q in the position q[2] = q(1) 
     s_array.append(s_incr)                                  #1.4
     t_incr = t_array[i-2] - (q_array[i]*t_array[i-1])     #1.5
        #Very important => Array q in the position q[2] = q(1)
     t_array.append(t_incr)                                  #1.5

    print("gcd(r0,r1) = r(i-1) = " + str(r_array[i-1]))       #2
    print("s = s(i-1) = " + str(s_array[i-1]))                #2
    print("t = t(i-1) = " + str(t_array[i-1]))                #2
    print("The inverse of " + str(r1) + " mod " + str(r0) + " is: " + str(t_array[i-1]))
```
