```python
for i in range(7):
    try:
     print(f"Inversion of {str(F(i))} mod 7 is " + str(F(i)**-1) + " mod 7")
    except Exception as e:
     print(f"An exception occurred: Inversion of {str(F(i))} mod 7 doesn't exist")
```

```python
from IPython.display import display
import pandas as pd
from collections import deque

i_array = deque()
j_array = deque()
mult_array = deque()

for i in range(7):
    for j in range(7):
        i_array.append(F(i))
        j_array.append(F(j))
        mult_array.append(F(i) * F(j))

# Create a sample DataFrame
df = pd.DataFrame({
    "F(i)": i_array,
    "F(j)": j_array,
    "F(i) x F(j) mod 7": mult_array,
})

# Display the DataFrame in a rich format
display(df)
```
