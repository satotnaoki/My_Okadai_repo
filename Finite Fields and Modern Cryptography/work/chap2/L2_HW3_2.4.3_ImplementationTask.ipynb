```python
from sage.all import *

def find_mult_order():
    p = 7
    F = GF(p) 
    for i in F:
        if i == 0: continue
        for k in range(1, p):
            if (i**k == 1): # i**k (mod p)                
                print("The order is " + str(k))
                print("for the element " + str(i))
                print(f"since {i}**{k} (mod {p}) = " + str(i**k) + "\n")
                break
```

```python
from sage.all import *
from sage.functions.log import logb
import random
import math


p = 7
F = GF(7)
F_pos = [n for n in F if n > 0]
print(f"GF({p}): " + str([n for n in F]))
print(f"GF({p})*: " + str(F_pos))

generators = [n for n in F if n.is_primitive_root()]

g = generators[0]
print("\nPicking the 1st generator (primitive element) g = " + str(g))

h = F(random.choice(F_pos))

print("Selecting random element h = " + str(h))
print(f"Calculating x = log (base g) of h = log (base {g}) of {h}")

print("\nWithout Sage library")
if (h == 1):
 x = 0
 print(f"x = log (base {g}) of {h} = " + str(x))
else:
 x = math.log(g, h)
 print(f"x = log (base {g}) of {h} = ", str(x)) 

print("\nWith Sage library")
x = logb(Integer(h), Integer(g))
print(f"x = log (base {g}) of {h} = ", str(x))

x = discrete_log(h,g)
print(f"x = log (base {g}) of {h} = ", str(x))
```

```python
print("Using brute force to find the exponent x. This example doesnt use any math logarithm library")

p = 7
g = 3 
h = pow(g,3,p)
print(f"h = {g}**3 mod {p} = " + str(h))
#x = math.log(g, h) 
#print(f"x = log (base {g}) of {h} = ", str(x)) 
value = 1
for x in range(p):  # x ranges from 0 to p-1   
 if value == h:
  print(f"The value of x, which gives {g}**x mod {p} = {h} is, " + str(x))
  break
 print(f" {value} (internal value) x {g} (g) mod {p}")    
 value = (value * g) % p
 print(f"  Equivalent to g**{x+1} mod 7, exponent (x = {x}) + 1")
 print(f"  {value} (internal value after mod {p})")    
```
