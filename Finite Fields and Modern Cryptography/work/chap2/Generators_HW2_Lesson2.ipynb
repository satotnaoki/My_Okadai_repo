```python
factors = divisors(abs(1008))
print(factors)
```

```python
from collections import deque
prime_factors = deque()


for n in factors:
 if n.is_prime():
    prime_factors.append(n)

prime_factors
```

```python
print("Another way to implement Euler's Totient function for p = 1009, and n = p - 1 = 1008 \n")

coprime_factors.clear()
coprime_factors = deque()

for i in range(1008):
     if (i % 2 != 0 and i % 3 != 0 and i % 7 != 0):
       #print(i+1)
       coprime_factors.append(i)

print(coprime_factors)
print("\nResult is matching the number of generators: " + str(len(coprime_factors)))
```
